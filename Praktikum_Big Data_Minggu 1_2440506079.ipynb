PRAKTIK PYSPARK LANJUTAN

```python
# Latihan 1: Memanipulasi Data dengan PySpark.
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

TUGAS 5 : Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

```python
# JAWABAN TUGAS 5 :
import pandas
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()


# Data PySpark
pyspark_data = [
    ("Ali", 34),
    ("Budi", 23),
    ("Citra", 29),
    ("Dina", 45)
]
pyspark_columns = ["Nama", "Usia"]
pyspark_df = spark.createDataFrame(pyspark_data, pyspark_columns)

print("PySpark:")
pyspark_df.show()


# Konversi data dari PySpark ke Pandas
df_pandas_1 = pyspark_df.toPandas()

print("Konversi PySpark ke Pandas:")
print(df_pandas_1)


# Data Pandas
data_pandas_2 = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Insinyur", "Progammer", "Dokter", "Guru"],
    "Hobi": ["Memancing", "Main Game", "Membaca", "Scroll Medsos"],
    "Gender": ["Laki-laki", "Laki-laki", "Perempuan", "Perempuan"]
}
df_pandas_2 = pandas.DataFrame(data_pandas_2)

print("\nPandas:")
print(df_pandas_2)


# Gabungkan df_pandas_1 dan df_pandas_2
df_joined = pandas.merge(df_pandas_1, df_pandas_2, on="Nama")

print("\nJoin:")
print(df_joined)


# Tampilkan df_joined menggunakan matplotlib
print("\n")
df_joined["Usia"].plot(kind="bar")
plt.show()
```

TUGAS 6 : Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

```python
# JAWABAN TUGAS 6
import pandas
import matplotlib.pyplot as plt

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()


# Data PySpark
pyspark_data = [
    ("Ali", 34),
    ("Budi", 23),
    ("Citra", 29),
    ("Dina", 45)
]
pyspark_columns = ["Nama", "Usia"]
pyspark_df = spark.createDataFrame(pyspark_data, pyspark_columns)

print("PySpark:")
pyspark_df.show()


# Konversi data dari PySpark ke Pandas
df_pandas_1 = pyspark_df.toPandas()

print("Konversi PySpark ke Pandas:")
print(df_pandas_1)


# Data Pandas
data_pandas_2 = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Insinyur", "Progammer", "Dokter", "Guru"],
    "Hobi": ["Memancing", "Main Game", "Membaca", "Scroll Medsos"],
    "Gender": ["Laki-laki", "Laki-laki", "Perempuan", "Perempuan"]
}
df_pandas_2 = pandas.DataFrame(data_pandas_2)

print("\nPandas:")
print(df_pandas_2)


# Gabungkan df_pandas_1 dan df_pandas_2
df_joined = pandas.merge(df_pandas_1, df_pandas_2, on="Nama")

print("\nJoin:")
print(df_joined)


# Operasi statistik
print("\nBeberapa operasi statistik dalam kolom \"Usia\"")
print(f"Tertinggi\t: {df_joined['Usia'].max()}")
print(f"Rata-rata\t: {df_joined['Usia'].mean()}")
print(f"Terendah\t: {df_joined['Usia'].min()}")
print(f"Median\t\t: {df_joined['Usia'].median()}")
print(f"Jumlah\t\t: {df_joined['Usia'].sum()}")
```
